```python
# Inicialização dos registradores
x10 = 1
x11 = 7
x12 = 0
x13 = 0
x15 = 2

# Memória simulada com lista
mem = [7, 7, 6]
```

```python
while True:
    x14 = mem[x12]

    if x14 != x11:
        break

    x13 = x14 + x10
    x12 += 1

mem[x15] = x13
```

```python
int main() {
    while (1) {
        x14 = mem[x12];

        if (x14 != x11) {
            break;
        }

        x13 = x14 + x10;
        x12 += 1;
    }
    
    mem[x15] = x13;
}
```
